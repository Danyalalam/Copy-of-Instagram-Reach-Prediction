```python
# prompt:  Let’s split the data into training and test sets before training the model:

import numpy as np
x = np.array(data[['Likes', 'Saves', 'Comments', 'Shares',
                   'Profile Visits', 'Follows']])
y = np.array(data["Impressions"])

xtrain, xtest, ytrain, ytest = train_test_split(x, y,
                                                test_size=0.2,
                                                random_state=42)
```

```python
# prompt: Now here’s is how we can train a machine learning model to predict the reach of an Instagram post using Python:

model = PassiveAggressiveRegressor()
model.fit(xtrain, ytrain)
y_pred = model.predict(xtest)

print("R2 score of the model is :",metrics.r2_score(ytest, y_pred))
```

```python
# prompt: Now let’s predict the reach of an Instagram post by giving inputs to the machine learning model:

import numpy as np
# Assuming user inputs for features
likes = 1000  #@param {type:"number"}
saves = 500  #@param {type:"number"}
comments = 100  #@param {type:"number"}
shares = 50  #@param {type:"number"}
profile_visits = 200  #@param {type:"number"}
follows = 10  #@param {type:"number"}

features = np.array([[likes, saves, comments, shares, profile_visits, follows]])
predicted_impressions = model.predict(features)

print("Predicted Impressions:", predicted_impressions[0])
```
